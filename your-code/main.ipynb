#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

plt.sca(axes[0])
monthly_median_steps = fitbit.groupby('Months_encoded')['Steps'].median()
monthly_median_steps.plot(kind='bar', color='pink')
plt.title('Median Steps by Month')
plt.xlabel('Month')
plt.ylabel('Median Steps')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
plt.sca(axes[1])  
weekday_vs_weekend = fitbit.groupby('Work_or_Weekend')['Steps'].mean()
weekday_vs_weekend.plot(kind='bar', color=['green', 'red'])
plt.title('Average Steps: Weekdays vs. Weekends')
plt.xlabel('Day Type')
plt.ylabel('Average Steps')
plt.xticks(range(2), ['Weekend', 'Weekday'])


import os
if not os.path.exists('figures'):
    os.makedirs('figures')

plt.savefig('figures/fitbit_visualization.png')


plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
combinations = [
    ('Minutes Lightly Active', 'Steps'),
    ('Minutes Very Active', 'Steps'),
    ('Minutes Sedentary', 'Steps')
]

fig, axes = plt.subplots(1, len(combinations), figsize=(15, 5))

for i, (feature1, feature2) in enumerate(combinations):
    plt.sca(axes[i])
    plt.scatter(fitbit[feature1], fitbit[feature2])
    plt.title(f'{feature1} vs {feature2}')
    plt.xlabel(feature1)
    plt.ylabel(feature2)

plt.show()
```
